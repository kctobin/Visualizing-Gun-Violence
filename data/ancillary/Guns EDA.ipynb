```python
import pandas as pd
import datetime as dt
import numpy as np

gva2014 = pd.read_json("./guns/scraper/src/gva/spiders/results/2014/2014.json")
gva2015 = pd.read_json("./guns/scraper/src/gva/spiders/results/2015/2015.json")
gva2016 = pd.read_json("./guns/scraper/src/gva/spiders/results/2016/2016.json")
gva2017 = pd.read_json("./guns/scraper/src/gva/spiders/results/2017/2017.json")
print gva2014.shape
print gva2015.shape
print gva2016.shape
print gva2017.shape


print gva2014.columns
print gva2015.columns
```

```python
comb = pd.concat([gva2014,gva2015,gva2016,gva2017])
comb = comb.reset_index(drop=True)
print comb.shape
```

```python
for col in comb:
    print col,": ", comb[col].isnull().sum()
```

```python
import types
victims = pd.DataFrame(columns = comb["victims"][0][0].keys())
for a in comb["victims"]:
    if type(a) == types.ListType:
        for b in a:
            vic = pd.DataFrame([b.values()], columns = b.keys())
            victims = pd.concat([vic,victims])

print victims.shape
```

```python
columns = comb.columns[[0,1,2,3,4,8]].values
columns = columns.tolist()
columns = columns + comb["victims"][0][0].keys()
victims2 = pd.DataFrame(columns = columns)
for a in range(len(comb)):
#for a in range(100):
    data1 = comb.iloc[a,[0,1,2,3,4,8]].tolist()
    if type(comb.loc[a]["victims"]) == types.ListType:
        for b in comb.loc[a]["victims"]:
            values = data1+b.values()
            vic = pd.DataFrame([values], columns = columns)
            victims2 = victims2.append(vic, ignore_index=True)

print victims2.shape
print victims2.head()
```
